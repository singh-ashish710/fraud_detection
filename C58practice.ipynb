## Bivariate Analysis
Numerical vs Numerical

```python
sns.scatterplot(x=df["Age"], y=df["Fare"])
plt.show()
```

```python
Correlation plot: df_corr = df.corr() # correlation
plot plt.figure(figsize = [8, 6])
sns.heatmap(df_corr, annot = True, fmt = '.1f', cmap = 'vlag_r', center = 0)
plt.title('Titanic dataset - correlation map\n', weight = 'bold', fontsize = 14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12);
```

## Numerical VS Categorical

```python
sns.boxplot(x=df["Pclass"], y=df["Age"])
plt.show()
```

```python
for col1 in cat_cols:
  for col2 in num_cols:
    sns.boxplot(x=df["col1"], y=df["col2"])
    plt.show()
```
